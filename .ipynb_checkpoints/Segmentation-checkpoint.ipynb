```python
import torch
from PIL import Image
from torchvision import transforms
import matplotlib.pyplot as plt
import numpy as np 
import cv2

# 텐서의 출력 제한 해제
torch.set_printoptions(threshold=torch.inf)
np.set_printoptions(threshold=np.inf)

torch.hub._validate_not_a_forked_repo=lambda a,b,c: True

# 이미지를 읽어서 original_image에 할당하는 예시 코드
# cv2_original_image = cv2.imread('../dog_breed/stanford/images/train/n02086740-닥스훈트(장모)/images - 2023-05-08T170538.320.jpg')
cv2_original_image = cv2.imread('./test4.jpg')
cv2_image = cv2.cvtColor(cv2_original_image, cv2.COLOR_BGR2RGB)
```

```python
# 입력 이미지 불러오기
original_image = Image.open('./test4.jpg')
input_image = original_image.convert("RGB")

# 입력 이미지를 딥러닝 모델에 넣을 수 있는 형태로 전처리 수행
preprocess = transforms.Compose([transforms.ToTensor(), transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225]),])
input_tensor = preprocess(cv2_original_image)
input_batch = input_tensor.unsqueeze(0)

#딥러닝 불러오기
model = torch.hub.load('pytorch/vision:v0.10.0', 'deeplabv3_resnet50', pretrained=True)
model.eval()

#딥러닝에 입력 이미지 넣고 예측결과 가져오기
with torch.no_grad():
    output = model(input_batch)['out'][0]
output_predictions = output.argmax(0)
# print(output_predictions)

segmentation_map = output_predictions.cpu().numpy()
# print(segmentation_map)

# max_value = np.max(segmentation_map)
# print(max_value)

# 세그멘테이션 맵의 각 클래스의 요소 수 계산
unique, counts = np.unique(segmentation_map, return_counts=True)
# unique = np.delete(unique,0)
class_counts = dict(zip(unique, counts))
class_counts = {k: v for k, v in class_counts.items() if k != 0}
print('class_counts: ',class_counts)

# 가장 많은 요소 수를 가진 클래스 찾기
max_count_class = max(class_counts, key=class_counts.get)
print('max_count_class: ',max_count_class)
```

```python
# 결과를 칼라로 표현하도록 색상표 만들기
palette = torch.tensor([2 ** 25 - 1, 2 ** 15 - 1, 2 ** 21 - 1])
colors = torch.as_tensor([i for i in range(21)])[:, None] * palette
colors = (colors % 255).numpy().astype("uint8")

# 결과 이미지 생성하기
r = Image.fromarray(output_predictions.byte().cpu().numpy()).resize(input_image.size)
# r.putpalette(colors)

# 결과 이미지 표출하기
plt.imshow(r)
plt.show()
```

```python
# 클래스 레이블에 해당하는 픽셀을 추출하여 세그멘테이션 마스크 생성
mask = np.zeros_like(segmentation_map)

# 해당 클래스의 모든 요소를 max_value로 설정
mask[segmentation_map == max_count_class] = 255


# 세그멘테이션 마스크 출력
print(mask)
```

```python
mask = mask.astype(np.uint8)

# 세그멘테이션 마스크로부터 객체의 경계를 찾음
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

# 경계를 기반으로 바운딩 박스 좌표를 계산
bounding_boxes = []
for contour in contours:
    x, y, w, h = cv2.boundingRect(contour)
    bounding_boxes.append((x, y, x+w, y+h))  # (top-left x, top-left y, bottom-right x, bottom-right y)

# 원본 이미지 위에 바운딩 박스를 표시
image_with_boxes = cv2_image.copy()

for bbox in bounding_boxes:
    x1, y1, x2, y2 = bbox
    cv2.rectangle(image_with_boxes, (x1, y1), (x2, y2), (0, 255, 0), 2)  # 바운딩 박스 그리기 (색상: 녹색, 두께: 2)
    
# 결과 이미지 출력
# cv2.imshow("Image with bounding boxes", image_with_boxes)
# cv2.waitKey(0)
# cv2.destroyAllWindows()


plt.imshow(cv2_image)
plt.axis('off')  # 축 제거
plt.show()

plt.imshow(image_with_boxes)
plt.axis('off')  # 축 제거
plt.show()
```

```python
# 바운딩 박스 좌표
x1 = x
y1 = y 
x2 = x+w 
y2 = y+h
print(x1, y1, x2, y2)

# 추출할 영역의 너비와 높이 계산
width = x2 - x1
height = y2 - y1
print(width, height)

# 영역 추출
extracted_area = cv2_original_image[y1:y2, x1:x2]

# # 추출된 영역을 새로운 이미지 파일로 저장
cv2.imwrite('./seg_images/extracted_area'+str(1)+'.jpg', extracted_area)


plt.imshow(Image.open('./seg_images/extracted_area1.jpg'))
plt.axis('off')  # 축 제거
plt.show()
```

```python
# 검정배경

# 세그멘테이션 마스크를 사용하여 개체 영역 추출
contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
object_mask = np.zeros_like(mask)
cv2.drawContours(object_mask, contours, -1, (255), thickness=cv2.FILLED)

# 개체 영역을 원본 이미지에서 자르기
object_image = cv2.bitwise_and(cv2_original_image, cv2_original_image, mask=object_mask)

# 잘린 개체 이미지 저장
cv2.imwrite('./seg_images/object_image.jpg', object_image)
```

```python
# 투명

contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
object_mask = np.zeros_like(mask)
cv2.drawContours(object_mask, contours, -1, (255), thickness=cv2.FILLED)

# 알파 채널 생성
alpha_channel = np.zeros_like(mask)
alpha_channel[object_mask != 0] = 255

# 알파 채널을 원본 이미지에 추가
object_image = cv2.cvtColor(cv2_original_image, cv2.COLOR_BGR2BGRA)
object_image[:, :, 3] = alpha_channel

# 잘린 개체 이미지 저장
cv2.imwrite('./seg_images/object_image.png', object_image)
```
